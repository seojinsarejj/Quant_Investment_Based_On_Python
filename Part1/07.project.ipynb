# 2. Load data

```python
import numpy as np
import pandas as pd

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

pd.set_option('display.float_format', lambda x : '%.3f' % x)
pd.set_option('max_columns', None)
```

```python
df = pd.read_csv("../inflearn_pandas_part1_material/my_data/fin_statement_new.csv")
df.head()
```

```python
df = df.drop(["상장일"], axis=1)
df = df.rename(columns={
    "DPS(보통주, 현금+주식, 연간)": "DPS",
    "P/E(Adj., FY End)": "PER",
    "P/B(Adj., FY End)": "PBR",
    "P/S(Adj., FY End)": "PSR",
})
```

```python
df.head()
```

# 3. yearly returns

```python
df[df["Name"] == "동화약품"]
```

```python
yearly_price_df = df.pivot(index="year",columns="Name", values="수정주가")
```

```python
yearly_price_df
```

```python
yearly_rtn_df = yearly_price_df.pct_change(fill_method=None).shift(-1) # 전년도 대비 수익률
yearly_rtn_df
```
